```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("/kaggle/input/pandas/friends.csv")
df.head()
```

```python
df.info()
```

```python
df.describe(include="all")
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=["Unnamed: 0"])
df.head()
```

```python
sns.histplot(df["marks"], kde=True)
plt.title("Distribuci√≥n de notas")
plt.show()
```

```python
sns.boxplot(x=df['marks'])
plt.title("Boxplot de notas")
plt.show()
```

```python
sns.countplot(x=df['city'])
plt.title("Distribuci√≥n de estudiantes por ciudad")
plt.show()
```

```python
sns.barplot(x=df['name'], y=df['marks'])
plt.title("Notas por estudiante")
plt.show()
```
